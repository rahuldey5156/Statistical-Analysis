```python
import numpy as np
import pandas as pd
import math
```

```python
d1 = pd.read_csv('monomer_CA.dat', header = None, delim_whitespace = True)
d2 = pd.read_csv('Monomer_CA_last2000ns_ChemShift_concat_datas.csv', header = None, delim_whitespace = True)

df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)

print(df1.shape)
print(df2.shape)
```

```python
data = []

for n in range(0, len(df1.iloc[0,:])):
  for i in range(0, len(df2.iloc[0,:])):
    sub = (df2.iloc[n, i] - df1.iloc[n, 0])
    square = sub ** 2
    rmsd = math.sqrt(square)
    data.append(rmsd)
```

```python
print(data)
```

```python
with open('rmsd_per_residue.dat', 'w') as file:
  for item in data:
    file.write("%s\n" % item)
```
