```python
import pandas as pd
import os
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
#Read all files
data1 = pd.read_csv('CA_F_18aver.dat', header=None)
data2 = pd.read_csv('CB_F_18aver.dat', header=None)
data3 = pd.read_csv('C_F_18aver.dat', header=None)

#Convert all files to DataFrame
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
```

```python
#Selecting columns
x1 = df1.iloc[:, 0]
x2 = df2.iloc[:, 0]
x3 = df3.iloc[:, 0]
```

```python
fig, ((ax1, ax2, ax3)) = plt.subplots(3, 1, gridspec_kw={'hspace': 0.36, 'wspace': 0.35})

#Histogram plot
ax1.hist(x1, bins=60, linewidth=0.5, edgecolor="black", color="blue", weights=np.ones_like(x1) / len(x1), alpha=0.7)
ax2.hist(x2, bins=60, linewidth=0.5, edgecolor="black", color="blue", weights=np.ones_like(x2) / len(x2), alpha=0.7)
ax3.hist(x3, bins=60, linewidth=0.5, edgecolor="black", color="blue", weights=np.ones_like(x3) / len(x3), alpha=0.7)

#Font Style
plt.rcParams.update({'font.sans-serif':'Times New Roman'})
csfont = {'fontname':'Times New Roman'}

#y Label and setting label location
ax1.set_ylabel(r'$C_{\alpha}$', labelpad=2, fontsize = 18, fontweight='bold')
ax2.set_ylabel(r'$C_{\beta}$', labelpad=2, fontsize = 18, fontweight='bold')
ax3.set_ylabel(r'C', labelpad=2, fontsize = 18, fontweight='bold')
ax3.set_xlabel(r'RMSD (ppm)', labelpad=2, fontsize = 18, fontweight='bold')

#Labelling all figures a, b, c, d
ax1.text(-0.29, 1.07, 'a)', ha='left', va='top', transform=ax1.transAxes,fontweight='bold', fontsize = 18)
ax2.text(-0.27, 1.06, 'b)', ha='left', va='top', transform=ax2.transAxes,fontweight='bold', fontsize = 18)
ax3.text(-0.29, 1.07, 'c)', ha='left', va='top', transform=ax3.transAxes,fontweight='bold', fontsize = 18)

# Increase the thickness of the axes
for spine in ax1.spines.values():
    spine.set_linewidth(1.5)

for spine in ax2.spines.values():
    spine.set_linewidth(1.5)

for spine in ax3.spines.values():
    spine.set_linewidth(1.5)

# Customize the thickness of ticks
ax1.tick_params(axis='both', which='major', width=2, labelsize=16)
ax1.tick_params(axis='both', which='minor', width=1, labelsize=16)

ax2.tick_params(axis='both', which='major', width=2, labelsize=16)
ax2.tick_params(axis='both', which='minor', width=1, labelsize=16)

ax3.tick_params(axis='both', which='major', width=2, labelsize=16)
ax3.tick_params(axis='both', which='minor', width=1, labelsize=16)

#Tick limits
ax1.set_xlim(3.32, 3.45)
ax3.set_xlim(3.32, 3.45)
ax1.set_ylim(0, 0.08)
ax2.set_ylim(0, 0.055)
ax3.set_ylim(0, 0.055)

# Save the plot
plt.tight_layout()
plt.savefig('RMSD_plot.png', dpi=500) 
plt.show()
```
