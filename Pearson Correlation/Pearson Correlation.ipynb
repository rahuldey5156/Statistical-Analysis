```python
import pandas as pd
import os
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the file
data = pd.read_csv('data.dat', header=None, delim_whitespace=True)
#Convert the file to DataFrame
df = pd.DataFrame(data)
```

```python
#Selecting columns
x = df.iloc[:, 0]
y = df.iloc[:, 1]
```

```python
#Calculating correlation and p value
corr, p_value = pearsonr(x, y)
print(f"Pearson correlation coefficient: {corr:.3f}")
print(f"P-value: {p_value:.3f}")
```

```python
fig, ax = plt.subplots()

#Scatter plot                  
ax.scatter(x, y, color ='black')

#Calculating slope and intercept for linear plot
slope = corr * (np.std(y) / np.std(x))
intercept = np.mean(y) - slope * np.mean(x)

#linear plot                   
plt.plot(x, slope * x + intercept, color='red')

#Font Style
plt.rcParams.update({'font.sans-serif':'Times New Roman'})
csfont = {'fontname':'Times New Roman'}

# Increase the thickness of the axes
for spine in ax.spines.values():
    spine.set_linewidth(1.5)

# Customize the thickness of ticks
ax.tick_params(axis='both', which='major', width=2, labelsize=16)

#Label and setting label location
ax.set_ylabel(r'$\delta$$_{\rm sim}$', labelpad=2, fontsize = 20, fontweight='bold')
ax.set_xlabel(r'$\delta$$_{\rm exp}$', labelpad=2, fontsize = 20, fontweight='bold')

plt.tight_layout()
plt.savefig('pearson_correlation.png', dpi=500)
plt.plot()
plt.show()
```
