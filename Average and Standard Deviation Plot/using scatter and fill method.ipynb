```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
data = pd.read_csv('datafile.dat', header=None, delim_whitespace=True)
df = pd.DataFrame(data)
```

```python
a = df.iloc[:,0]
x = df.iloc[:,1]*10
y = df.iloc[:,2]*10
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,7))
ax.fill_between(a, x-y, x+y, color='black', alpha=0.2)
ax.scatter(a, x, color='black', s=20, label='trimer')
ax.plot(a, x, linestyle='-', color='black')

#Font Style
plt.rcParams.update({'font.sans-serif':'Times New Roman'})
csfont = {'fontname':'Times New Roman'}

#Axis Label and setting label location
ax.set_xlabel(r'Residue index', labelpad=2, fontsize = 30)
ax.set_ylabel(r'RMSF ($\AA$)', labelpad=2, fontsize = 30)

# Increase the thickness of the axes
for spine in ax.spines.values():
    spine.set_linewidth(2)

# Customize the thickness of ticks
ax.tick_params(axis='x', which='major', length=5, width=2, labelsize=0)
ax.tick_params(axis='y', which='major', length=5, width=2, labelsize=30)
ax.tick_params(axis='both', which='minor', length=5, width=1, labelsize=30)

#hline
ynew1 = 4
ynew2 = 8
ax.axhline(ynew1, linestyle='--', color='black', linewidth=0.25)
ax.axhline(ynew2, linestyle='--', color='black', linewidth=0.25)

#Tick limits
ax.set_xlim(306, 378)
ax.set_ylim(0, 10)

#Tick range
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.yaxis.set_major_locator(MultipleLocator(4))

plt.tight_layout()
plt.savefig('scatter_avg_fill_std.png', dpi=300) 
plt.show()
```
