```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
data = pd.read_csv('datafile.dat', header=None, delim_whitespace=True)
df = pd.DataFrame(data)
```

```python
a = df.iloc[:,0]
x = df.iloc[:,1]*10
y = df.iloc[:,2]*10
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,7))

ax.plot(a, x, color='blue', label='trimer')
ax.fill_between(a, x-y, x+y, color='blue', alpha=0.4)

#Font Style
plt.rcParams.update({'font.sans-serif':'Times New Roman'})
csfont = {'fontname':'Times New Roman'}

#Axis Label and setting label location
ax.set_xlabel(r'Residue index', labelpad=2, fontsize = 20)
ax.set_ylabel(r'RMSF ($\AA$)', labelpad=2, fontsize = 20)

# Increase the thickness of the axes
for spine in ax.spines.values():
    spine.set_linewidth(2)

# Customize the thickness of ticks
ax.tick_params(axis='x', which='major', length=5, width=2, labelsize=20)
ax.tick_params(axis='y', which='major', length=5, width=2, labelsize=20)
ax.tick_params(axis='both', which='minor', length=5, width=1, labelsize=20)

ynew1 = 4
ynew2 = 8
ax.axhline(ynew1, linestyle='--', color='black', linewidth=0.25)
ax.axhline(ynew2, linestyle='--', color='black', linewidth=0.25)

#Tick limits
ax.set_xlim(272, 381)
ax.set_ylim(0, 12)

#Tick range
ax.xaxis.set_major_locator(MultipleLocator(20))
ax.yaxis.set_major_locator(MultipleLocator(4))

plt.tight_layout()
plt.savefig('avg_std_using_fill_between.png', dpi=300) 
plt.show()
```
